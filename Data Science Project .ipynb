```python
specializations = ['Web Development', 'Machine Learning', 'Network Security', 'Data Science', 'Artificial Intelligence']
counts = [402902, 40142, 39932, 39870, 39764]

plt.figure(figsize=(9, 7))
plt.bar(specializations, counts, color= ['skyblue','teal','coral','tan','orange'])

for i in range(len(specializations)):
    plt.text(i, counts[i] + 1000, str(counts[i]), ha='center')

plt.xlabel('Specialization')
plt.ylabel('Count')
plt.title('Count of Specializations')
plt.show()
```

```python
plt.figure(figsize=(11, 8))
field_of_study_counts = df["Field of Study"].value_counts()
colors = ['tan', 'teal', 'cyan', 'coral', 'orange']
plt.bar(field_of_study_counts.index, field_of_study_counts.values, color=colors)

for i, count in enumerate(field_of_study_counts.values):
    plt.text(i, count + 50, str(count), ha='center', va='bottom', fontsize=10)

plt.xlabel("Field of Study")
plt.ylabel("Number of Students")
plt.title("Distribution of Students by Field of Study")
plt.show()
```

```python

df['Date of Birth'] = pd.to_datetime(df['Date of Birth'])

current_year = datetime.now().year
df['age'] = current_year - df['Date of Birth'].dt.year

plt.hist(df['age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Students')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
df['Year of Admission'] = pd.to_datetime(df['Year of Admission'])

admissions_per_year = df.groupby(df['Year of Admission'].dt.year).size()

plt.figure(figsize=(10, 6))
plt.plot(admissions_per_year.index, admissions_per_year.values, marker='o')
plt.title('Trend of Student Admissions Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Admissions')
plt.grid(True)
plt.xticks(admissions_per_year.index) 
plt.show()
```

```python
grouped_data = df.groupby(['Field of Study', 'Specialization']).size().unstack()
grouped_data.plot(kind='bar', stacked=True)

plt.title('Distribution of Students Across Specializations Within Each Field of Study')
plt.xlabel('Field of Study')
plt.ylabel('Number of Students')
plt.legend(title='Specialization')   
plt.show()
```
